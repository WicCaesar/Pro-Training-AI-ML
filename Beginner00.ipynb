```python
# Read the data from the CSV file with read from pandas
data = pd.read_csv("Salary_dataset.csv");

# Configure the style
sns.set_theme(style="darkgrid", palette="deep", font="sans-serif", color_codes='true');

# Plot the data points with seaborn
ex00 = sns.jointplot(data=data, x='YearsExperience', y='Salary', color='m', kind="scatter", height=7);
#ex00.fig.suptitle("Aumento salarial", y=1.02, color="purple");
#ex00.set_axis_labels("Anos de experiência", "Salário");
#ex00.ax_joint.set_xlabel("Anos de experiência", color="purple");
#ex00.ax_joint.set_ylabel("Salário", color="green");

#* Joint plots show the relationship between two variables along with their individual distributions.
```

```python
# We're going to use this library and find out how it works on the Internet
from sklearn.linear_model import LinearRegression;

# Separate the features (YearsExperience) from the target variable (Salary)
#X = data[['YearsExperience']]; #* The double square brackets ensure that X is a DataFrame (the expected input format for many Machine Learning libraries).
X = data['YearsExperience']; #* In Machine Learning, X tipically represents ~~features~~ or ~~independent variables~~ that will be used to make predictions.
y = data['Salary']; #* And y tipically represents the ~~target~~ or ~~dependent variable~~ that you want to predict.

# Create a linear regression model
regression = LinearRegression(); #* This function comes from the sklearn.linear_model module.

# Train the model
#train = regression.fit(X, y);
train = regression.fit(X.values.reshape(-1, 1), y);
#* Reshape means to change the shape of an array without changing its data.
#* The fit method adjusts the model's parameters to minimize the difference between the predicted values and the actual values in y.
#* The reshape(-1, 1) is used to convert the 1D array X into a 2D array with one column, which is the expected input format for the fit method.
#* The -1 in reshape means "unspecified", allowing NumPy to automatically determine the appropriate number of rows based on the length of the original array.

# Predict the Salary values based on YearsExperience
#predictions = regression.predict(X);
predictions = regression.predict(X.values.reshape(-1, 1));
```
